```python
def missing_values_table(df):
  mis_val = 100 * df.isnull().sum() / len(df)
  mis_val = mis_val.sort_values()
  mis_val = mis_val[mis_val > 0]
  return mis_val
```

```python
numeric = df.select_dtypes(exclude='object')
```

```python
objet = df.select_dtypes(include='object')
```

```python
dummies = pd.get_dummies(objet, drop_first=True)
```

```python
df_final = pd.concat([numeric, dummies], axis=1)
```

```python
df_final.head()
```

```python
X = df_final.drop('SalePrice', axis=1)
y = df_final['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scale_X_train = scale.fit_transform(X_train)
```

```python
scale_X_test = scale.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(scale_X_train, y_train)
```

```python
y_pred = model_lr.predict(scale_X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_lr = mean_squared_error(y_test, y_pred)
```

```python
mse_lr
```

```python
rmse_lr = np.sqrt(mse_lr)
```

```python
rmse_lr
```

```python
from sklearn.linear_model import ElasticNet
```

```python
model_el = ElasticNet(random_state=42)
```

```python
param_grid = {'l1_ratio': np.linspace(0, 1, 3),
              'alpha': [0.01, 0.1, 1, 10, 100],
              'max_iter': [10000]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_el = GridSearchCV(model_el, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
```

```python
grid_el.fit(scale_X_train, y_train)
```

```python
grid_el.best_params_
```

```python
y_pred = grid_el.predict(scale_X_test)
```

```python
mse_el = mean_squared_error(y_test, y_pred)
```

```python
rmse_el = np.sqrt(mse_el)
```

```python
print(mse_el)
print(rmse_el)
```

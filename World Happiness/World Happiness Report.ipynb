**Essai de constitution en un seul Dataset**

```python
'''
Calcul de standard Error à partir de la formule (Upper - Lower)/2*z.
Étant donné qu'on à considéré 90 pourcent comme niveau de confiance,
vu que c'est petit dataset alors le z est pris à 1.645.
'''
df_2016['Standard Error'] = (df_2016['Upper Confidence Interval'] - df_2016['Lower Confidence Interval'])/2*1.645
```

```python
#Suppression des variables Upper et Lower
df_2016 = df_2016.drop(['Lower Confidence Interval', 'Upper Confidence Interval'], axis=1)
```

```python
df_2016.head()
```

```python
#Changement de position pour la colonne Standard Error pour qu'elle corresponde à celui du dataset 2015
col = df_2016.pop('Standard Error')
df_2016.insert(4, 'Standard Error', col)
```

```python
df_2016.head()
```

```python
#Création d'une colonne Year pour spécifier les années.
df_2015['Year'] = 2015
df_2016['Year'] = 2016
```

```python
df_2015_2016 = pd.concat([df_2015, df_2016])
```

```python
df_2015_2016.head()
```

```python
df_2017['Standard Error'] = (df_2017['Whisker.high'] - df_2017['Whisker.low'])/2*1.645
df_2017 = df_2017.drop(['Whisker.high', 'Whisker.low'], axis=1)
df_2017['Year'] = 2017
#Remplacer les valeurs manquantes au niveau de Région pour le dataset de 2017 par Na.
df_2017['Region'] = pd.NA
```

```python
df_2017.head()
```

```python
#J'ordonne les colonne pour les faire correspondre à celui de 2015 pour faciliter une concaténation plutard
df_2017 = df_2017[['Country', 'Region', 'Happiness.Rank', 'Happiness.Score', 'Standard Error', 'Economy..GDP.per.Capita.',
'Family', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.', 'Generosity', 'Dystopia.Residual', 'Year']]
```

```python
#Écrire de la même manière les noms des colonnes de 2015 à 2017.
df_2017.columns = df_2015.columns
```

```python
df_2017.head()
```

```python
#Concaténation des dataset de 2015, 2016, 2017
df_2015_a_2017 = pd.concat([df_2015_2016, df_2017])
```

```python
df_2015_a_2017.head()
```

```python
df_2018.head()
```

```python
#Création de la colonne Year pour 2018
df_2018['Year'] = 2018
```

```python
df_2019.head()
```

```python
#Création de la colonne Year pour 2019
df_2019['Year'] = 2019
```

```python
#Concaténation des datasets 2018 et 2019 car ils ont les mêmes variables
df_2018_2019 = pd.concat([df_2018, df_2019])
```

```python
df_2018_2019.head()
```

```python
#Remplacer les valeurs manquantes au niveau de la colonne Perceptions of corruption par la médiane
df_2018_2019['Perceptions of corruption'] = df_2018_2019['Perceptions of corruption'].fillna(df_2018_2019["Perceptions of corruption"].median())
```

```python
df_2018_2019.head()
```

```python
df_2015_a_2017.head()
```

```python
#Créer les variables Region, Dystopia Residual, Standard Error pour permettre au dataset df_2018_201 d'avoir les meêmes variables
df_2018_2019['Region'] = pd.NA
df_2018_2019['Dystopia Residual'] = pd.NA
df_2018_2019['Standard Error'] = pd.NA
```

```python
#Ranger les variables dans l'ordre du dataset de 2015
df_2018_2019 = df_2018_2019[['Country or region', 'Region', 'Overall rank', 'Score', 'Standard Error', 'GDP per capita',
                             'Social support', 'Healthy life expectancy', 'Freedom to make life choices',
                             'Perceptions of corruption', 'Generosity', 'Dystopia Residual', 'Year']]
```

```python
#Faire coreespondre les noms
df_2018_2019.columns = df_2015_a_2017.columns
```

```python
#Concaténation finale
df = pd.concat([df_2015_a_2017, df_2018_2019])
```

```python
#Dataset final
df.head()
```

```python
#Pourcentage de valeur manquante
100*df.isnull().sum()/len(df)
```

```python
#Rempli la les valeurs manquantes de la colonne Region en groupant sur Country
df.groupby('Country')['Region'].transform('first')
```

```python
df['Region'] = df['Region'].fillna(df.groupby('Country')['Region'].transform('first'))
```

```python
100*df.isnull().sum()/len(df)
```

```python
#Remplacer les valeurs manquantes au niveau de Standard Error par la médiane
df['Standard Error'] = df['Standard Error'].fillna(round(df['Standard Error'].median(), 5))
```

```python
100*df.isnull().sum()/len(df)
```

```python
#Remplacer les valeurs manquantes au niveau de Dystopia Residual en tenant compte du Country
df['Dystopia Residual'] = df['Dystopia Residual'].fillna(df.groupby('Country')['Dystopia Residual'].transform('median'))
```

```python
100*df.isnull().sum()/len(df)
```

```python
#Remplacer le reste des valeurs manquantes au niveau de Region par Unknown, c-à-d Inconnu.
df['Region'] = df['Region'].fillna('Unknown')
```

```python
100*df.isnull().sum()/len(df)
```

```python
#Supprimer le reste des lignes contenants de valeurs manquantes
df = df.dropna()
```

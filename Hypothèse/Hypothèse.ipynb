```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, ttest_ind, chi2_contingency, fisher_exact
```

```python
df['restecg'] = df['restecg'].fillna(df['restecg'].mode()[0])
```

```python
df['thalch'] = df['thalch'].fillna(df['thalch'].median())
```

```python
df['exang'] = df['exang'].fillna(df['exang'].mode()[0])
```

```python
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].median())
```

```python
df['slope'] = df['slope'].fillna(df['slope'].mode()[0])
```

```python
df['ca'] = df['ca'].fillna(df['ca'].median())
```

```python
df['thal'] = df['thal'].fillna(df['thal'].mode()[0])
```

```python
df = df.drop('id', axis=1)
```

```python
df.head()
```

#Shapiro-Wilk

Hypothèse:
Ho: La variable âge suis une loi normale

```python
homme = df[df['sex'] == 'Male']
```

```python
femme = df[df['sex'] == 'Female']
```

```python
stats, p = shapiro(df['age'])
```

```python
p
```

La variable âge ne suis pas une loi normale car p <= 0.05 donc Ho est rejétéé

#Test d'indépendance

```python
stats, p = ttest_ind(homme['age'], femme['age'])
```

```python
p
```

Pas de preuve que l'âge a d'effet sur le fait d'aoir la maladie ou pas

#Khi-deux

Hypothèse:
1.   Ho: La variable sexe et slope sont liées
2.   H1: La variable sexe et slope ne sont pas liées

```python
stats, p, dof, expected = chi2_contingency(pd.crosstab(df['sex'], df['slope']))
```

```python
p
```

p < 0.05 alors l'hypothèse Ho est accepté et H1 est rejeté. On peut donc conclu que les variables sex et slope sont liées.

#Fisher

```python
stats, p = fisher_exact(pd.crosstab(df['sex'], df['slope']))
```

```python
p
```

Ceci viens de confirmer encore plus l'hypothèse du test de Khi-deux
